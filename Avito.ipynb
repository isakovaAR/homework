### Найдте общее количество запросов, которое написал аналитик

В среднем, на 1 тяжелый запрос уходит 7/60 условной памяти, на 1 легкий запрос в среднем уходит 5/60. Понятно, что на 1 "стандартный" запрос не может уйти больше (меньше) памяти, чем в среднем на тяжелый (легкий) запрос.Таким образом, верно:

5/60 < x/20 < 7/60,

а поскольку x есть целое число, то единственным подходящим вариантом является: х = 2. Т.е. на 1 "стандартный" процесс уходит 1/10 от всей памяти. Заметим, что всего памяти осталось 8/20, а значит всего "стандартных" процессов будет 8*10/20 = 4 (а всего 10)

2. Один из аналитиков дал неправильный ответ, тогда менеджер выбирает наугад

# Задача 3

Пусть А - это событие, что на втором подбрасывании выпадет число, меньше 4 (т.е. 3, 2, или 1), 
С - это событие, что при первом броске выпало 4.
Необходимо найти условную вероятность P(A|C) = P(AC)/P(C).

Посчитаем для начала знаменатель по формуле полной вероятности: P(C) = P(C на 1 кубике)Р(выбрали 1 кубик) + Р(С на 2 кубике)Р(выбрали 2 кубик) + Р(С на 3 кубике)Р(выбрали 3 кубик) = 1/3 * (1/6 + 1/12 + 1/20) = 1/10,
здесь за скобкой 1/3, т.к. любой кубик можно вытянуть равновероятно.

Теперь посчитаем числитель, который также по формуле полной вероятности распадется уже на 3! = 6 слагаемых: Р(АС) = Р(С на 1 кубике, А на 2 кубике)Р(выбрали для первого броска 1 кубик)Р(выбрали для второго броска 2 кубик)+ ... = 1/3 * 1/2 * (1/6 * 3/12 + ... ) =  19/720,
здесь вынесли за скобку 1/6, т.к. на 1 броске вероятность вытянуть любой кубик всегда 1/3, а на 2 броске всегда равна 1/2.

Итог, Р(А|C) = 19/720 : 1/10 = 19/72 

## Ответ: 19/72

# Задача 4

## Предложенные метрики

1. Количество активных пользователей в месяц.

Если проект действительно успешен, то между пользователями Авито возникнет больше доверия, люди начинают больше пользоваться сайтом и советовать его своим знакомым

2. Коэффициент оттока пользователей. 

Возможна и обратная ситуация, если например люди решают приобрести какой-нибудь товар, основываясь на ложных отзывах, то велик риск мошенничества. Это, в свою очередь, может привести к потери клиента. Будет наблюдаться отток.

3. Частота использования данной услуги.

Если количество отзывов сильно не возрастает, или люди практически не пользуются рейтинговой системой, то стоит задуматься, в чем может быть проблема. Например на некоторых сайтах до сих пор необходимо регестрироваться, чтобы оставить отзыв (к Авито это не относится), что пользователи могут находить неудобным или даже опасным.

С другой стороны, если люди охотно начинают оставлять отзывы, взаимодействовать между собой и непосредственно с продавцом,то это показатель качетсва проекта. Это говорит о том, что сайт стал удобнее для пользователей, поскольку, ориентируясь на отзывы,  они могут выбрать лучший товар из предлагаемых и поделиться своими впечатлениями 

4. Количесвто людей, зарегестрировавшихся на сайте.

Если после начала проекта этот показатель начал расти, то это говорит о том, что пользователи начали больше доверять Авито. Зарегестрированные пользователи чаще всего те люди, которые находят сайт самым удобным и практичным для использования, это его регулярные посетители, и если их становится все больше, то проект успешен. 

5. Относительная доля "застоявшихся" объявлений.

Главная цель данного проекта - позволить клиентам оценить качетсво покупаемого ими товара до момента покупки. Если введенная система действительно прогрессирует, то посетители начнут все больше обращать внимания на товары, у которых есть отзывы и высокий рейтинг, и все меньше на товары без отзывов. Это может привести к застою объявлений, когда на них никто не откликается. Однако с точки зрения проекта, это хороший показатель его эффективности 

6. Коэффициент продаж у продавцов с высоким и низким рейтингом.

Правильное и эффективное использование данной системы должно привести к изменению уровня продаж разных продавцов. Те, у кого выскоий рейтинг и хорошие отзывы увеличивают свои продажи, а те, у кого низкий рейтинг, наоборот их теряют

## Как оценить долю фальшивых отзывов?

1. Например, одной из отличительных черт настоящего отзыва (в частности положительного) является тот факт, что после его появления на сайте другие клиенты начинают чаще покупать товары у данного продавца и оставлять такие же положительные отзывы. В случае, когда отзыв ложен, ситуация складывается другим образом, покупатели начинают оставлять негативные отзывы (чаще всего даже с подчеркиванием того, что их обманули). Если заблаговременно отслеживать подобное поведение и затем запретить продавцу выкладывать объявления на Авито, то это может предупредить большую цепочку мошенничеств


2. В случае, когда мы хотим определить является ли чей-то негативный отзыв клеветой со стороны конкурентов, можно отследить активность аккаунта, который оставил этот отзыв. Если раннее клиент не смотрел номер телефона продавца, или в его поиске не было названия данного продукта, или у него почти нет активности (т.е. нет просмторов других телефонов, нет истории поиска и т.д.), то можно сделать вывод, что отзыв не настоящий.


3. Аналогично можно поступить и с ложными положительными отзывами, то есть если у автора отзыва нет активности до этого момента (включая историю поиска и отстуствие просмотра телефона продавцва), то скорее всего кто-то намерено заходит чтобы оставить положительный отзыв и привлечь внимание покупателей.


4. Также чтобы вычилить накрутку можно посмотреть на оценки к комментариям (если оценки есть). Если система из людей, которые оставили отзывы и оценили другие, замкнута, то это говорит о намеренном увеличении рейтинга.


5. Наличие отзывов от клиентов с большой активностью (т.е. с большой историей покупок, поиска и т.д.). Это в свою очередь является гарантом качества товара и говорит о том, что продавцу нет нужды накручивать себе рейтинг


Все эти методы можно использовать для того, чтобы отличить истинные отзывы от ложных, и вычислить долю накрутки у того или иного продавца
